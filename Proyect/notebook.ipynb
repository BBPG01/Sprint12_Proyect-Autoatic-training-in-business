<B> DEFINIR FUNCION COMPLETA

```python
def train_and_evaluate(data):
    
    features = data[['f0', 'f1', 'f2']]
    target = data['product']
    
    features_train, features_valid, target_train, target_valid = train_test_split(
        features,
        target,
        test_size=0.25,
        random_state=12345
    )
    
    model = LinearRegression()
    model.fit(features_train, target_train)
    
    predictions_valid = model.predict(features_valid)
    
    rmse = mean_squared_error(target_valid, predictions_valid, squared=False)
    mean_prediction = predictions_valid.mean()
    
    return rmse, mean_prediction
```

Por lo tanto, el siguiente paso consiste en:

seleccionar los 200 pozos con las mayores predicciones de reservas en cada región,

calcular el beneficio esperado basado en estas selecciones,

y evaluar el riesgo de pérdidas mediante simulaciones (bootstrap) antes de tomar una decisión final.

<b>ESCRIBIR UNA FUNCION PARA CALCULAR LA GANACIA DE UN VONJUNTO DE POZOS DE PETROLEO SELECCIONADOS (MODELADO DE PREDICCIONES)

<b>Función para entrenar, predecir y devolver resultados

```python
def get_predictions(data):
    
    features = data[['f0', 'f1', 'f2']]
    target = data['product']
    
    features_train, features_valid, target_train, target_valid = train_test_split(
        features,
        target,
        test_size=0.25,
        random_state=12345
    )
    
    model = LinearRegression()
    model.fit(features_train, target_train)
    
    predictions = model.predict(features_valid)
    
    return pd.Series(predictions), target_valid.reset_index(drop=True)
```

<b>Funcion para calcular ganacia

```python
def calculate_profit(predictions, target):
    top_predictions = predictions.sort_values(ascending=False).head(N_WELLS)
    real_reserves = target[top_predictions.index]
    total_reserves = real_reserves.sum()
    profit = total_reserves * REVENUE_PER_UNIT - BUDGET
    return profit, total_reserves
```

<b>Ejecutar para cada region

```python
regions = {
    'Región 0': reg_1,
    'Región 1': reg_2,
    'Región 2': reg_3
}
results = {}

for region, data in regions.items():
    predictions, target_valid = get_predictions(data)
    profit, reserves = calculate_profit(predictions, target_valid)
    
    results[region] = {
        'profit': profit,
        'reserves': reserves
    }
```

<b>Mostrar resultados

```python
for region, values in results.items():
    print(region)
    print('Reservas totales:', values['reserves'])
    print('Ganancia potencial ($):', values['profit'])
```

<b>CALCULO DE RIESGOS Y GANANCIAS PARA CADA REGION

<b>Constantes

```python
BOOTSTRAP_SAMPLES = 1000
```

<b>Funcion para calcular ganacias

```python
def calculate_profit(predictions, target):
    top_predictions = predictions.sort_values(ascending=False).head(N_WELLS)
    real_reserves = target[top_predictions.index]
    profit = real_reserves.sum() * REVENUE_PER_UNIT - BUDGET
    return profit
```

<b>Funcion del bootstrapping

```python
def bootstrap_profit(predictions, target):
    profits = []
    
    for _ in range(BOOTSTRAP_SAMPLES):
        subsample = predictions.sample(n=len(predictions), replace=True, random_state=None)
        subsample_target = target[subsample.index]
        
        profit = calculate_profit(subsample, subsample_target)
        profits.append(profit)
    
    return pd.Series(profits)
```

<b>Ejecutarlo para cada region

```python
bootstrap_results = {}

for region, data in regions.items():
    predictions, target_valid = get_predictions(data)
    profits = bootstrap_profit(predictions, target_valid)
    
    bootstrap_results[region] = profits
```

<b>Calculo de metricas finales

```python
for region, profits in bootstrap_results.items():
    mean_profit = profits.mean()
    ci_lower = profits.quantile(0.025)
    ci_upper = profits.quantile(0.975)
    loss_risk = (profits < 0).mean() * 100
    
    print(region)
    print('Ganancia promedio:', mean_profit)
    print('IC 95%:', (ci_lower, ci_upper))
    print('Riesgo de pérdidas (%):', loss_risk)
```

<b>Conclusiones

Los resultados muestran que las tres regiones presentan un riesgo de pérdidas del 0%, lo que indica una alta estabilidad del modelo al seleccionar los 200 pozos con mayor volumen predicho.

Sin embargo, al comparar las ganancias promedio y los intervalos de confianza, se observan diferencias claras:

Región 0 presenta la mayor ganancia promedio (≈ 166.4 millones de dólares) y el intervalo de confianza más alto.

Región 2 ocupa el segundo lugar en ganancia promedio (≈ 152.9 millones de dólares).

Región 1 muestra la ganancia promedio más baja (≈ 148.2 millones de dólares).

Dado que la Región 0 combina la mayor ganancia esperada con un riesgo de pérdidas nulo, se propone esta región como la mejor candidata para el desarrollo de pozos petrolíferos.
